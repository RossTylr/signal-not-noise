# 01a — Building Intuition: What Is a Dimension?

## The One-Sentence Version
A dimension is just a question you can ask about something — and every answer needs its own axis to live on.

## What You'll Build Intuition For
- What a "dimension" actually means outside of spatial metaphors
- Why we can't visualise past 3D but can still reason about higher dimensions
- How to think about 10,000-dimensional data without your brain melting

## Prerequisites
- None — this is where we start

## The Story

Imagine you're a doctor. A patient walks in and you measure their age: 34. That single number lives on a number line — one dimension.

Now you check their heart rate: 72 beats per minute. You need a second axis to record this, so you've moved into two dimensions. Plot them on a flat scatter plot — age on one axis, heart rate on the other. One dot, two questions answered.

You clip on a pulse oximeter and get blood oxygen: 97%. Now you need a third axis. We can still picture this — it's a point floating in a 3D box you can rotate around.

But then you take blood pressure. And respiratory rate. And run a blood panel with 20 values. Suddenly you've got 25 numbers describing this one patient. Each number is a dimension. Each dimension is just *a question you asked*.

You can't picture 25 dimensions. Nobody can. But that doesn't matter — the maths works exactly the same whether you have 3 dimensions or 25 or 25,000. A photo from a phone camera? That's a point in a space with *millions* of dimensions (one per pixel channel). A genome? Twenty thousand dimensions of gene expression.

The goal of this notebook is to make that feel *normal*. By the end, "100-dimensional data" should sound as unremarkable as "a spreadsheet with 100 columns" — because that's exactly what it is.

## A Dimension Is a Question

Let's start with a single patient and a single question. One question = one number = one dimension.

We'll ask three questions, one at a time, and watch the space grow from a number line to a 3D world you can rotate.

```python
# --- One dimension: a single question ---
# Patient A, age 34. One number → one axis → a number line.

fig, ax = plt.subplots(figsize=(10, 2))
ax.scatter([34], [0], s=120, color=COLOURS["signal"], zorder=5)
ax.annotate("Patient A\nage = 34", (34, 0), textcoords="offset points",
            xytext=(0, 18), ha="center", fontsize=11, color=COLOURS["signal"])
ax.set_xlim(10, 70)
ax.set_ylim(-0.5, 0.8)
ax.set_xlabel("Age")
ax.set_yticks([])
ax.set_title("One dimension: one question, one axis")
ax.axhline(0, color=COLOURS["neutral"], linewidth=0.8, alpha=0.5)
plt.tight_layout()
plt.savefig("visuals/01a_one_dimension.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# --- Two dimensions: a second question ---
# Now we also know heart rate = 72. Two numbers → two axes → a flat plane.

fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter([34], [72], s=120, color=COLOURS["signal"], zorder=5)
ax.annotate("Patient A\n(34, 72)", (34, 72), textcoords="offset points",
            xytext=(15, 12), ha="left", fontsize=11, color=COLOURS["signal"])
ax.set_xlabel("Age")
ax.set_ylabel("Heart Rate (bpm)")
ax.set_xlim(10, 70)
ax.set_ylim(50, 100)
ax.set_title("Two dimensions: two questions, two axes")
plt.tight_layout()
plt.savefig("visuals/01a_two_dimensions.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# --- Three dimensions: a third question ---
# Blood oxygen (SpO2) = 97%. Three numbers → three axes → a 3D space.
# This one is interactive — drag to rotate.

fig_3d = go.Figure(data=[go.Scatter3d(
    x=[34], y=[72], z=[97],
    mode="markers+text",
    marker=dict(size=8, color=COLOURS["signal"]),
    text=["Patient A (34, 72, 97)"],
    textposition="top center",
    textfont=dict(size=12)
)])

fig_3d.update_layout(
    title="Three dimensions: three questions, three axes",
    scene=dict(
        xaxis_title="Age",
        yaxis_title="Heart Rate (bpm)",
        zaxis_title="SpO₂ (%)",
        xaxis=dict(range=[10, 70]),
        yaxis=dict(range=[50, 100]),
        zaxis=dict(range=[90, 100]),
    ),
    width=700, height=500,
    margin=dict(l=0, r=0, t=40, b=0),
)
fig_3d.show()
```

> **Key insight:** A "dimension" isn't mystical. It's just another column in your spreadsheet.
> 10,000 dimensions = 10,000 columns. Each column is a question you asked about your data.

## Adding More Patients

One patient is a single dot. Let's add 49 more and see how data fills up 1D, 2D, and 3D space.

We'll generate 50 synthetic patients with age, heart rate, and blood oxygen — and we'll bake in a deliberate correlation between age and heart rate (older patients tend to have slightly higher resting heart rates). Watch for it.

```python
# Generate 50 synthetic patients
rng = np.random.default_rng(42)
n_patients = 50

age = rng.normal(45, 15, n_patients)
heart_rate = 0.3 * age + rng.normal(70, 10, n_patients)  # slight correlation with age
blood_ox = rng.normal(97, 2, n_patients)                  # mostly independent

patients = np.column_stack([age, heart_rate, blood_ox])
print(f"Shape: {patients.shape}")   # (50, 3) — 50 patients, 3 dimensions
print(f"That's {patients.shape[0]} patients described by {patients.shape[1]} measurements each.")
```

```python
# Side-by-side: how the same 50 patients look in 1D, 2D, and 3D

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- 1D: strip plot (jittered) ---
jitter = rng.uniform(-0.15, 0.15, n_patients)
ax1.scatter(age, jitter, s=40, alpha=0.6, color=COLOURS["signal"], edgecolors="white", linewidth=0.5)
ax1.set_xlabel("Age")
ax1.set_yticks([])
ax1.set_ylim(-0.5, 0.5)
ax1.set_title("1D: Ages on a number line")
ax1.axhline(0, color=COLOURS["neutral"], linewidth=0.5, alpha=0.4)

# --- 2D: scatter plot ---
ax2.scatter(age, heart_rate, s=40, alpha=0.6, color=COLOURS["signal"], edgecolors="white", linewidth=0.5)
ax2.set_xlabel("Age")
ax2.set_ylabel("Heart Rate (bpm)")
ax2.set_title("2D: Age vs Heart Rate")

plt.tight_layout()
plt.savefig("visuals/01a_50_patients_1d_2d.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# --- 3D interactive: all three measurements ---

fig_3d_multi = go.Figure(data=[go.Scatter3d(
    x=age, y=heart_rate, z=blood_ox,
    mode="markers",
    marker=dict(size=4, color=COLOURS["signal"], opacity=0.7),
    text=[f"Age={a:.0f}, HR={h:.0f}, SpO₂={s:.1f}" for a, h, s in zip(age, heart_rate, blood_ox)],
    hoverinfo="text",
)])

fig_3d_multi.update_layout(
    title="3D: 50 patients in Age × Heart Rate × SpO₂ space",
    scene=dict(
        xaxis_title="Age",
        yaxis_title="Heart Rate (bpm)",
        zaxis_title="SpO₂ (%)",
    ),
    width=700, height=500,
    margin=dict(l=0, r=0, t=40, b=0),
)
fig_3d_multi.show()
```

**Did you notice?** In the 2D scatter, age and heart rate trend upward together. Older patients tend to have higher resting heart rates — the two measurements aren't fully independent.

That upward trend means these two dimensions are *partially redundant*. They're both encoding some of the same underlying information. We'll dig into this in **Module 02: Redundancy and Structure**, where we'll see that redundancy is actually the reason dimensionality reduction works at all.

## What Happens at Dimension 4?

You can't see 4D. Nobody can. But that doesn't stop us from *working* with it.

We'll add a fourth measurement — blood pressure — and use two tricks to peek into 4D:
1. **Colour-coding**: use colour to represent the 4th dimension on a 3D plot
2. **Pair plots**: show every possible 2D slice of the data side-by-side

```python
# Add a 4th dimension: blood pressure (correlated with age)
bp = 0.5 * age + rng.normal(120, 15, n_patients)

# Trick 1: 3D scatter with colour = 4th dimension
fig_4d = go.Figure(data=[go.Scatter3d(
    x=age, y=heart_rate, z=blood_ox,
    mode="markers",
    marker=dict(
        size=5,
        color=bp,
        colorscale="Viridis",
        colorbar=dict(title="BP (mmHg)"),
        opacity=0.8,
    ),
    text=[f"Age={a:.0f}, HR={h:.0f}, SpO₂={s:.1f}, BP={b:.0f}"
          for a, h, s, b in zip(age, heart_rate, blood_ox, bp)],
    hoverinfo="text",
)])

fig_4d.update_layout(
    title="4D via colour: BP encoded as point colour in 3D space",
    scene=dict(
        xaxis_title="Age",
        yaxis_title="Heart Rate (bpm)",
        zaxis_title="SpO₂ (%)",
    ),
    width=700, height=500,
    margin=dict(l=0, r=0, t=40, b=0),
)
fig_4d.show()
```

```python
# Trick 2: Pair plot — every 2D slice of 4D data
# 4 dimensions → 6 unique pairs, shown as a 4×4 scatter matrix

patients_4d = np.column_stack([age, heart_rate, blood_ox, bp])
labels = ["Age", "HR", "SpO₂", "BP"]

fig, axes = plt.subplots(4, 4, figsize=(12, 12))
for i in range(4):
    for j in range(4):
        ax = axes[i, j]
        if i == j:
            ax.hist(patients_4d[:, i], bins=15, alpha=0.7,
                    color=COLOURS["signal"], edgecolor="white")
        else:
            ax.scatter(patients_4d[:, j], patients_4d[:, i],
                       alpha=0.5, s=20, color=COLOURS["signal"],
                       edgecolors="white", linewidth=0.3)
        if i == 3:
            ax.set_xlabel(labels[j])
        else:
            ax.set_xticklabels([])
        if j == 0:
            ax.set_ylabel(labels[i])
        else:
            ax.set_yticklabels([])

fig.suptitle("Pair plot: every 2D slice of our 4D patient data", fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig("visuals/01a_pair_plot.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"4 dimensions → {4 * 3 // 2} unique 2D projections (the off-diagonal plots).")
```

> **Key insight:** We can't see high dimensions, but we can slice them, project them, and measure them. The maths doesn't care that our eyes stop at 3.
>
> Look at the pair plot: notice how Age-HR and Age-BP show clear correlations (upward trends), while SpO₂ looks independent of everything else. We're already reading 4D structure from 2D slices.

## Dimensions as Degrees of Freedom

There's another way to think about dimensions: they're the **independent ways something can vary**.

A physicist would call these *degrees of freedom*. The more ways a system can change, the higher-dimensional the space you need to describe it.

| System | Degrees of Freedom | Why |
|--------|-------------------|-----|
| Train on a track | 1 | Forward/backward only |
| Boat on a lake | 2 | North/south, east/west |
| Drone in the air | 3 | Add up/down |
| Drone with camera gimbal | 4 | Add camera angle |
| Casualty evacuation scenario | 20+ | Casualties, severity, helicopters, distance, weather, threats... |

In a simulation like FAER-M, a casualty evacuation scenario might have dozens of dimensions: number of casualties, severity distribution, helicopter availability, distance to Role 2, weather conditions, route threats... Each is an axis in the scenario space.

The "curse" we'll meet in **01b** explains why you can't just grid-search this space.

## The Leap to Thousands

So far we've worked with 3-4 dimensions. That's cosy. Real data is often *massively* higher-dimensional.

Let's look at a classic example: handwritten digits. Each digit is an 8×8 greyscale image — which means each image is a point in **64-dimensional space** (one dimension per pixel).

```python
# Load the sklearn digits dataset via our utility
data, target, images = make_digit_data()

print(f"Each digit image is a point in {data.shape[1]}D space")
print(f"We have {data.shape[0]} such points")
print(f"Image shape: {images[0].shape} → flattened to a {data.shape[1]}-element vector")

# Show a digit as an image alongside its 64-element vector
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.imshow(images[0], cmap="gray_r")
ax1.set_title("What we see: an image", fontsize=13)
ax1.set_xticks([])
ax1.set_yticks([])

ax2.bar(range(64), data[0], color=COLOURS["signal"], alpha=0.7, edgecolor="white", linewidth=0.3)
ax2.set_title("What the maths sees: a point in 64D space", fontsize=13)
ax2.set_xlabel("Dimension (pixel index)")
ax2.set_ylabel("Brightness")

plt.tight_layout()
plt.savefig("visuals/01a_digit_dimensions.png", dpi=150, bbox_inches="tight")
plt.show()
```

And 64 dimensions is tiny by modern standards. Here's what real high-dimensional data looks like:

| Domain | Dimensions | What Each Dimension Is |
|--------|-----------|----------------------|
| Genomics | ~20,000 | Gene expression levels |
| Medical imaging (100×100) | 10,000 | Pixel brightness |
| NLP (word embeddings) | 300–1,536 | Semantic feature |
| Patient records (EHR) | 500–5,000 | Lab values, diagnoses, medications |
| Simulation parameters | 10–100+ | Scenario configuration variables |

> **Key insight:** There is nothing special about 3 dimensions except that our eyes evolved to see them. The maths that works in 2D works in 20,000D. But — and this is the **but** — weird things start happening in high dimensions. That's what we'll explore in **01b: The Curse of Dimensionality**.

<details>
<summary><b>The Maths Behind This</b></summary>

Formally, when we say data is $d$-dimensional, we mean each data point is a vector in $\mathbb{R}^d$:

$$\mathbf{x} = (x_1, x_2, \ldots, x_d) \in \mathbb{R}^d$$

A dataset of $n$ points forms an $n \times d$ matrix $\mathbf{X}$, where each row is a data point and each column is a feature (dimension).

The **ambient dimension** is $d$ — the number of columns. But the data might not actually *use* all $d$ directions. If two columns are perfectly correlated, one is redundant — the data really lives in $d-1$ dimensions. The true number of independent directions the data varies along is called the **intrinsic dimension**, and it's almost always much less than $d$.

This gap between ambient and intrinsic dimension is the entire reason dimensionality reduction works.

</details>

## Where This Matters

**Healthcare:** Every patient encounter generates a high-dimensional data point. A single hospital admission might record hundreds of lab values, vital signs over time, medication administrations, imaging results, and clinical notes. When a clinician says "this patient looks like that patient," they're implicitly doing dimensionality reduction — compressing hundreds of features into a gut-level similarity judgement. The methods in this repo make that process explicit and rigorous.

**Simulation and Operations Research:** A casualty evacuation scenario in FAER-M isn't a single number — it's a point in a 20+ dimensional space. Understanding how that space behaves (and misbehaves) is the difference between running 50 useful scenarios and running 5,000 that teach you nothing new.
